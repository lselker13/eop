# Puneet

```python
puneet_metadata = pd.read_csv(
    '/Users/leo/Documents/gpl/eop/overseen_work/puneet/25SmrInt/output/albania_metadata.csv'
)
```

```python
puneet_metadata_edited = pd.read_csv(
    '/Users/leo/Documents/gpl/eop/overseen_work/puneet/25SmrInt/output/albania_metadata_edited.csv'
)
```

```python
puneet_metadata
```

```python
differences = puneet_metadata.compare(puneet_metadata_edited)
rows_with_differences = differences.index

for idx in rows_with_differences:
    row_original = puneet_metadata.loc[idx]
    row_edited = puneet_metadata_edited.loc[idx]
    combined_row = row_original.copy()
    
    for col in differences.columns.levels[0]:
        if differences.at[idx, (col, 'self')] != differences.at[idx, (col, 'other')]:
            combined_row[col + '_original'] = differences.at[idx, (col, 'self')]
            combined_row[col + '_edited'] = differences.at[idx, (col, 'other')]
            combined_row.drop(col, inplace=True)
    
    combined_row.name = idx
    combined_row_df = combined_row.to_frame().T
    
    if 'result_df' in locals():
        result_df = pd.concat([result_df, combined_row_df], axis=0)
    else:
        result_df = combined_row_df

display(result_df)
```
